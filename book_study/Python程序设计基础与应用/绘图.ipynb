```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ("Frogs", "Hogs", "Dogs", "Logs")
colors = ("#FF0000", "yellowgreen", "gold", "blue")
explode = (0, 0.02, 0, 0.08)
```

```python
fig = plt.figure(num=1, figsize=(10, 8), dpi=110, facecolor="white")
ax = fig.gca()
ax.pie(
    np.random.random(4),
    explode=explode,
    labels=labels,
    colors=colors,
    pctdistance=0.8,
    autopct="%1.1f%%",
    shadow=True,
    startangle=90,
    radius=0.25,
    center=(0, 0),
    counterclock=False,
    frame=True,
)
ax.pie(
    np.random.random(4),
    explode=explode,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    shadow=True,
    startangle=45,
    radius=0.25,
    center=(1, 1),
    frame=True,
)
ax.pie(
    np.random.random(4),
    explode=explode,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    shadow=True,
    startangle=90,
    radius=0.25,
    center=(0, 1),
    frame=True,
)
ax.pie(
    np.random.random(4),
    explode=explode,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    shadow=False,
    startangle=135,
    radius=0.35,
    center=(1, 0),
    frame=True,
)
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])
ax.set_xticklabels(["Sunny", "Cloudy"])
ax.set_yticklabels(["Dry", "Rainy"])
ax.set_xlim((-0.5, 1.5))
ax.set_ylim((-0.5, 1.5))
ax.set_aspect("equal")
plt.show()
```

```python
# 

import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

mpl.rcParams["legend.fontsize"] = 10  # 

fig = plt.figure()
ax = fig.gca(projection="3d")  # 
theta = np.linspace(-4 * np.pi, 4 * np.pi, 200)
z = np.linspace(-4, 4, 200)  # 
r = z**3 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label="parametric curve")
ax.legend()
plt.show()
```
